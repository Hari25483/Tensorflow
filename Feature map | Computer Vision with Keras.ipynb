```python
import keras
from keras import layers
import cv2
```

```python
img=cv2.imread('cat.jpeg')
img=cv2.resize(img,(224,224))
cv2.imshow('cat',img)
key=cv2.waitKey(0)
if key == ord("q"):
    print("Quit")
cv2.destroyAllWindows()

```

```python
import numpy as np
img=cv2.imread('cat.jpeg',cv2.IMREAD_GRAYSCALE)
img=cv2.resize(img,(224,224))
model=keras.Sequential()
model.add(layers.Conv2D(input_shape=(224,224,1),filters=64,kernel_size=(3,3)))
feature_map=model.predict(np.array([img]))
print(feature_map.shape)
```

```python
import matplotlib.pyplot as plt
feature_img=feature_map[0,:,:,1]
plt.imshow(feature_img)
plt.show()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
for i in range(64):
    feature_img=feature_map[0,:,:,i]
    ax=plt.subplot(8,8,i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.imshow(feature_img)
plt.show()
```

```python
img=cv2.imread('cat.jpeg')
img=cv2.resize(img,(224,224))
b,g,r=cv2.split(img)
for i in range(3):
    ax=plt.subplot(1,3,i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    if (i==1):
        plt.imshow(r)
    elif (i==2):
        plt.imshow(g)
    else:
        plt.imshow(b)

plt.show()
```
